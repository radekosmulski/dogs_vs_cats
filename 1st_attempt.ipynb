```python
import os, shutil, random, glob
```

```python
os.listdir('data')
```

```python
if os.path.exists('data/test'):
    shutil.rmtree('data/test')
os.makedirs('data/test/unknown')
    
for file in glob.glob('test/*'):
    shutil.copy(file, 'data/test/unknown')
```

```python
def create_dir(dir_name):
    # deletes a specified directory under 'data/train' and recreates
    # directory structure suitable for reading by keras
    if os.path.exists('data/train/' + dir_name):
        shutil.rmtree('data/train/' + dir_name)
    os.makedirs('data/train/' + dir_name)
    os.makedirs('data/train/' + dir_name + '/cats')
    os.makedirs('data/train/' + dir_name + '/dogs')

all_files = glob.glob('train/*')

# create full train set
create_dir('full')

cats = [f for f in all_files if 'cat' in f]
dogs = [f for f in all_files if 'dog' in f]

for category in ['cats', 'dogs']:
    for f in eval(category):
        shutil.copy(f, os.path.join('data/train/full/' + category, f.split('/')[1]))
```
